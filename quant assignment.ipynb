```python
x=data.drop(['datetime','signal'],axis=1)
```

```python
y=data['signal']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
xtrain_scaled = sc.fit_transform(xtrain)
xtest_scaled = sc.transform(xtest)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
xtrain_pca = pca.fit_transform(xtrain_scaled)
xtest_pca = pca.transform(xtest_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain_pca, ytrain)
```

```python
y_pred_train = model.predict(xtrain_pca)
y_pred_test = model.predict(xtest_pca)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
train_accuracy = accuracy_score(ytrain, y_pred_train)
test_accuracy = accuracy_score(ytest, y_pred_test)
```

```python
print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
print("Classification Report for Test Set:")
print(classification_report(ytest, y_pred_test))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100], 
    'penalty': ['l1', 'l2'], 
    'solver': ['liblinear'] 
}
```

```python
lr_classifier = LogisticRegression()
```

```python
grid_search = GridSearchCV(estimator=lr_classifier, param_grid=param_grid, cv=3, scoring='accuracy')
```

```python
grid_search.fit(x,y)
```

```python
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model = LogisticRegression(**best_params)
best_model.fit(xtrain, ytrain)
```
